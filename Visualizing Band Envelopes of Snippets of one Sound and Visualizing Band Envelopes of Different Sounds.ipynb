## Visualizing Band Envelopes of Snippets of one Sound and Visualizing Band Envelopes of Different Sounds

<font size="3">We will try to visualize each band envelope of different snippets of one sounds and we will visualize band envelopes of different sounds. This might help us recognize why certain marginal statistics have different values between snippets of one sound or between completely different sounds.</font>

```python
import os
import sys
sys.path.append('..')
from scipy.io import wavfile
import settings
import soundanalysis
import sound_files
from matplotlib import pyplot as plt
import numpy as np
import scipy
from scipy import stats
```

<font size="3">Read the first snippet of the sound heavy_rain and visualize each band envelope.</font>

```python
fig, axs = plt.subplots(2,2)

filename2 = os.path.join(settings.SOUNDS_PATH, 'heavy_rain_snippet1.wav')
sound2 = soundanalysis.SoundAnalysis(soundfile=filename2)
#(nSamples, samplingRate, soundDuration) = soundanalysis.get_sound_parameters()
filterbank_sound2 = soundanalysis.FilterBank(4, [20, 8000], len(sound2.wave), sound2.samplingRate)
tf_sound2 = filterbank_sound2.get_transfer_function()
(tvec, bandsWaveforms) = sound2.apply_filterbank(tf_sound2)
(tvecEnv, bandsEnvelopes) = sound2.calculate_bands_envelopes()

samplesToPlot = np.arange(2000,3000)
bandToPlot = 0
axs[0,0].plot(tvec[samplesToPlot],bandsWaveforms[bandToPlot,samplesToPlot])
axs[0,0].plot(tvecEnv[samplesToPlot],bandsEnvelopes[bandToPlot,samplesToPlot], lw=2)

samplesToPlot = np.arange(2000,3000)
bandToPlot = 1
axs[0,1].plot(tvec[samplesToPlot],bandsWaveforms[bandToPlot,samplesToPlot])
axs[0,1].plot(tvecEnv[samplesToPlot],bandsEnvelopes[bandToPlot,samplesToPlot], lw=2)

samplesToPlot = np.arange(2000,3000)
bandToPlot = 2
axs[1,0].plot(tvec[samplesToPlot],bandsWaveforms[bandToPlot,samplesToPlot])
axs[1,0].plot(tvecEnv[samplesToPlot],bandsEnvelopes[bandToPlot,samplesToPlot], lw=2)

samplesToPlot = np.arange(2000,3000)
bandToPlot = 3
axs[1,1].plot(tvec[samplesToPlot],bandsWaveforms[bandToPlot,samplesToPlot])
axs[1,1].plot(tvecEnv[samplesToPlot],bandsEnvelopes[bandToPlot,samplesToPlot], lw=2)
```
