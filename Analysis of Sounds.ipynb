## Analysis of Sounds

<font size="3">We will analyze three different sounds in this notebook. We will plot their waveforms, spectra, spectrograms, create a filterbank and calculate the marginal statistics of each sound. <br>Let's start by reading the three different sounds and plotting their waveforms.</font>

```python
import os
import sys
sys.path.append('..')
import settings
import soundanalysis
import scipy.io.wavfile as wavfile
from matplotlib import pyplot as plt
import numpy as np

filename = os.path.join(settings.SOUNDS_PATH, 'storm_clipped.wav')
sana = soundanalysis.SoundAnalysis(soundfile=filename)

frequency, data = wavfile.read(filename)
# check dimensionality of data
print(data.shape)
print(data)
# it gives number of samples, 
# if frequency is how many samples happen in one second
# then length of our "time" x-axis will be number of samples over frequency
length = data.shape[0]/frequency
print(length)
# and we get the length to be 6s, which is how much we clipped our song exactly
time = np.linspace(0,length,data.shape[0])
amplitude = data[:,]
print(amplitude)

plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(time, amplitude, color="magenta")

```

```python
import os
import sys
sys.path.append('..')
import soundanalysis
import settings
from matplotlib import pyplot as plt
import numpy as np

filename = os.path.join(settings.SOUNDS_PATH,'weir_in_flood_clipped.wav')
sana = soundanalysis.SoundAnalysis(soundfile=filename)


fbins = np.fft.fftfreq(len(sana.wave), 1/sana.samplingRate)
ff = np.fft.fft(sana.wave)


plt.plot(fbins, np.abs(ff), color="magenta")
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```

<font size="3"> Lastly, calculate marginal statistics of each band's envelope in each one of the sounds.</font>

```python
import os
import sys
sys.path.append('..')
import settings
from matplotlib import pyplot as plt
import soundanalysis
from soundanalysis import FilterBank

filename = os.path.join(settings.SOUNDS_PATH,'weir_in_flood_clipped.wav')
sana = soundanalysis.SoundAnalysis(soundfile=filename)

fbank = FilterBank(6, [20,8000], len(sana.wave), sana.samplingRate)
fullTF = fbank.get_transfer_function()
sana.apply_filterbank(fullTF)

sana.calculate_bands_envelopes()
sana.apply_compression(0.3)

bandStats = sana.calculate_bands_stats()
sana.plot_bands_stats()
```
